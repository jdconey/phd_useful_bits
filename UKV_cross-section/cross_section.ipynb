# Met Office UKV Cross-Section Plotter 

The Met Office's high-resolution UKV model is run every hour and forecasts a wide variety of variables. 
Sometimes you want a quick look at a cross-section of some of these data.
This particular notebook uses the 'level 1' processed netCDF files (regridded onto a regular 2km grid at 33 vertical levels above sea level rather than terrain following), though I'm sure with some quick tweaking you'd adapt it to your uses. 

It needs 2 files:

1. The data you want to plot, e.g. vertical velocities, or cloud amount on heights. Basically some 3D data (x,y,z)

2. The orography heights, so you can see where your variable of interest relates to the terrain - this doesn't need to be from the same model run as the data you want to plot, unless there's been something a bit drastic.

If you want to plot towns and cities, you need the `towns.csv` somewhere handy. But if not, set `add_towns` to `False`.
